```python
my_list = ["Jack", "Ahmed", "Jose","Amit"]

index = 0
slen = len(my_list)
print(f"Slen is {slen}")
while index < len(my_list):
    print(f"Index is {index}")
    print(my_list[index])
    index += 1
```

```python
creature = {"Name":"Alwin", "type":"ChipMunk", "age":2, "Furryness":"High"}
i = 0
while i < len(creature):
    key, value = list(creature.items())[i]
    print(f"{key}- {value}")
    i += 1
```
